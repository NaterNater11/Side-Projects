```python
import math
import time
import random
import numpy as np
import matplotlib.pyplot as plt



def prime(amount):
    primes = [2]
    num = 3
    even = [2]
    doubles = 0

    #loop
    while num <= amount:

        #reset cycling checker
        check = 0

        #effiecent way to check all possible divisions
        while primes[check] < math.ceil(math.sqrt(num)):

            #number can be divided, not prime break loop
            if num % primes[check] == 0:
                break

            #checking next prime number
            check += 1

        #if successfully ran through prime checker, put number into a list
        if num % primes[check] != 0:
            primes.append(num)


        if num % 2 == 0:
            #quick  way to check some even numbers
            if num/2 == primes[doubles]:
                doubles += 1
                even.append(num)

            #this would calculate two different primes that add to even numbers
            else:
                low = 1
                high = len(primes)-1
                while high != low:

                    #prime match for even number
                    if primes[high] + primes[low] == num:
                        even.append(num)
                        break

                    elif primes[high] + primes[low] < num:
                        low += 1

                    elif primes[high] + primes[low] > num:
                        high -= 1

        num += 1


def timer(times):

    t1 = time.perf_counter()
    prime(times)
    t2 = time.perf_counter()
    return (t2 - t1) * 1000.0

points = {}
cycles = 10

x_axis = []
y_axis = []

while cycles <= 10**6:
    
    x_axis.append(cycles)
    points[cycles] = (timer(cycles))
    y_axis.append(points[cycles])
    cycles *= 10
    
plt.loglog(x_axis, y_axis, 'bo-')
plt.xlabel('Number Amount')
plt.ylabel('Time [ms]')
plt.title("Time to Add Two Primes to Equal a Even Numbers")
plt.show()
```
