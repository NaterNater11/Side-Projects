Finding a Pair of Prime Numbers That Add to a Selected Even Number

```python
import math

#loop
def Value(n):
    add = {4:[2,2], 6:[3,3]}
    primes = [2]
    num = 3
    even = [2]
    doubles = 0
    while num <= n:

        #reset cycling checker
        check = 0

        #effiecent way to check all possible divisions
        while primes[check] < math.ceil(math.sqrt(num)):

            #number can be divided, not prime break loop
            if num % primes[check] == 0:
                break

            #checking next prime number
            check += 1

        #if successfully ran through prime checker, put number into a list
        if num % primes[check] != 0:
            primes.append(num)


        if num % 2 == 0:

            low = 1
            high = len(primes)-1
            while high != low:

                #prime match for even number
                if primes[high] + primes[low] == num:
                    add[num] = [primes[low], primes[high]]
                    break

                    #Smaller number of pair needs to be raised
                elif primes[high] + primes[low] < num:
                    low += 1


                    #Larger number of pair needs to be raised
                elif primes[high] + primes[low] > num:
                        high -= 1


        num += 1
    return(add)
    

num = 20
if num % 2 == 0:
    collection = Value(num)
    print(collection[num])
    
else:
    print('error: requires even number')
```
